# Social Media Engagement Analysis

## Introduction

This project analyzes social media engagement data to understand how different factors affect user interaction, including platform, post type, posting day, and sentiment.

## Data Loading

```python
import pandas as pd
df = pd.read_csv('social_media_engagement1.csv')
```

## Data Overview

```python
df.info()
df.isnull().sum()
df.describe()
```

## Engagement by Platform

```python
platform_engagement = df.groupby('platform')[['likes','comments','shares']].mean()
platform_engagement
```

```python
import matplotlib.pyplot as plt

platform_engagement.plot(kind='bar')
plt.title('Average Engagement by Platform')
plt.ylabel('Average Count')
plt.show()
```

### Insight

Instagram shows the highest engagement across likes, comments, and shares, while Twitter has the lowest.

This suggests that visually-driven platforms tend to generate more interaction from users.

## Engagement by Post Type

```python
post_type_engagement = df.groupby('post_type')[['likes','comments','shares']].mean()
post_type_engagement
```

```python
post_type_engagement.plot(kind='bar')
plt.title('Engagement by Post Type')
plt.ylabel('Average Count')
plt.show()
```

### Insight

Video and poll posts generate higher engagement compared to text and image posts.

This indicates that interactive and dynamic content is more effective in capturing user attention.

## Engagement by Day of Week

```python
day_engagement = df.groupby('post_day')[['likes','comments','shares']].mean()
day_engagement
```

```python
day_engagement.plot(kind='bar')
plt.title('Engagement by Day of Week')
plt.ylabel('Average Count')
plt.show()
```

### Insight

Engagement tends to be higher on weekends, especially Sunday.

This suggests that users are more active and responsive during their free time.

## Engagement by Sentiment

```python
sentiment_engagement = df.groupby('sentiment_score')[['likes','comments','shares']].mean()
sentiment_engagement
```

```python
sentiment_engagement.plot(kind='bar')
plt.title('Engagement by Sentiment')
plt.ylabel('Average Count')
plt.show()
```

### Insight

Posts with negative sentiment receive the highest engagement.

This may be because emotionally charged or controversial content attracts more user reactions.

## Conclusion

This analysis reveals several key patterns:

- Instagram achieves the highest engagement, highlighting the effectiveness of visual platforms
- Video and poll content perform better than other post types
- Engagement is generally higher on weekends, especially Sunday
- Negative sentiment posts tend to attract more interaction, possibly due to stronger emotional responses

These insights can help guide social media strategies to maximize user engagement.
